```python
import numpy as np
import librosa
import matplotlib.pyplot as plt

audio_path = "FÃ¼r Elise Performed by Lang Lang.mp3"
y, sr = librosa.load(audio_path, sr=None)

```

```python
frame_length_sec = 0.1      # 100 ms
hop_length_sec = 0.01       # 10 ms
fmin = librosa.note_to_hz('A0')  # Start at A0

hop_length_samples = int(hop_length_sec * sr)
bin_per_octave = 36
n_bins = 267

hop_length_samples
CQT = librosa.cqt(y=y, sr=sr, hop_length=hop_length_samples, bins_per_octave=bin_per_octave, n_bins=n_bins, fmin=fmin)
CQT_db = librosa.amplitude_to_db(CQT, ref=np.max)

librosa.display.specshow(CQT_db, sr=sr, x_axis='time', y_axis='cqt_note' )
plt.colorbar(format='%+2.0f dB')
plt.title(f'CQT Spectrogram of 0.01')
plt.xlabel('Time (s)')
plt.ylabel('CQT Frequency')
plt.show()
```

```python
frame_length_sec = 0.1  # 100 ms
hop_length_sec = 0.01  # 10 ms
fmin = librosa.note_to_hz('A0')  # Start at A0

hop_length_samples = int(hop_length_sec * sr)
bin_per_octave = 36
n_bins = 267

HCQT = librosa.hybrid_cqt(y=y, sr=sr, hop_length=hop_length_samples, bins_per_octave=bin_per_octave, n_bins=n_bins, fmin=fmin)
HCQT_db = librosa.amplitude_to_db(HCQT, ref=np.max)

plt.figure(figsize=(10, 6))
librosa.display.specshow(HCQT_db, sr=sr, x_axis='time', y_axis='cqt_note', cmap='inferno')

plt.colorbar(label='dB')
plt.title('Hybrid CQT Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```
